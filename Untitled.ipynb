```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
config_file = "frozen_inference_graph.pb"
frozen_model = "ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt"
```

```python
model = cv.dnn_DetectionModel(frozen_model,config_file)
```

```python
labels = []
file = 'labels.txt'
with open(file,'rt') as fpt:
    labels = fpt.read().rstrip('\n').split('\n')
labels
```

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5)
model.setInputMean(127.5)
model.setInputSwapRB(True)
```

```python
img = cv.imread('images\sksk.png')
plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
```

```python
ClassIndex, confidence, bbox = model.detect(img,confThreshold = 0.55)
```

```python
font_scale = 3
imgt = img
font = cv.FONT_HERSHEY_PLAIN
for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten() ,bbox):
    cv.rectangle(imgt,boxes,(250,0,0),2)
    cv.putText(imgt,labels[ClassInd-1],(boxes[0]+10,boxes[1]+40), font, fontScale = font_scale, color = (0,255,0),thickness = 3)
```
